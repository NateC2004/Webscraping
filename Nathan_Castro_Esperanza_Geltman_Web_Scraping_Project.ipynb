```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
# making list
Name_list = []
ID_list = []
Title_list = []
Type_list = []
Location_list = []
Subject_list = []
Deadline_list = []
Description_list = []

for url in list_of_urls:
  # for missing value
  Name= ''
  ID= ''
  Title= ''
  Type= ''
  Location= ''
  Subject= ''
  Deadline= ''
  link = ''

  response = requests.get(url)
  soup = bs(response.content)
  name = soup.title.text
  grid = soup.find('div', class_= "grid2 nopd")
  # Check if grid is found before proceeding
  if grid:
    divs_list = grid.find_all('div')
    for i in range(0,len(divs_list),2):
      key = divs_list[i].text
      if 'ID' in key:
        ID = divs_list[i+1].text
      if 'Title' in key:
        Title = divs_list[i+1].text
      if 'Type' in key:
        Type = divs_list[i+1].text
      if 'Location' in key:
        Location = divs_list[i+1].text
      if 'Subject' in key:
        Subject = divs_list[i+1].text
      if 'Deadline' in key:
        Deadline = divs_list[i+1].text
      if 'Description' in key:
        a_tag = divs_list[i+1].find('a')
        if a_tag:
          link = a_tag['href']

  Name_list.append(name)
  ID_list.append(ID)
  Title_list.append(Title)
  Type_list.append(Type)
  Location_list.append(Location)
  Subject_list.append(Subject)
  Deadline_list.append(Deadline)
  Description_list.append(link)

data_dictionary = {'Name':Name_list,
                   'ID':ID_list,
                   'Title':Title_list,
                   'Type':Type_list,
                   'Location':Location_list,
                   'Subject':Subject_list,
                   'Deadline':Deadline_list,
                   'Application Link':Description_list}
df = pd.DataFrame(data_dictionary)
df
```
